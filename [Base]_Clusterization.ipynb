````python
def plot_2d_data(data, labels, title=' ', cmap='tab20', ax=None):
    '''
     2d scatter plot.
    :param np.ndarray data: 2d  
    :param Union[list, np.ndarray] labels:      
    :param str title:  
    :param str cmap:  
    :param ax Optional[matplotlib.axes.Axes]:    .
           ,          
        ,     .    
    '''
    n_clusters = len(np.unique(labels))

    if ax is None:
        fig, ax = plt.subplots(1, 1, figsize=(10, 5))
    else:
        fig = None

    scatter = ax.scatter(
        data[:, 0], data[:, 1], c=labels,
        cmap=plt.get_cmap(cmap, n_clusters)
    )

    cbar = plt.colorbar(scatter, label=' ', ax=ax)
    cbar.set_ticks(np.min(labels) + (np.arange(n_clusters) + 0.5) * (n_clusters - 1) / n_clusters)
    cbar.set_ticklabels(np.unique(labels))

    ax.set_title(title)
    ax.grid(True)

    if fig is not None:
        fig.tight_layout()
        plt.show()
````

````python
plot_2d_data(data, labels, title=' ', cmap='tab20')
````
